```python
lb=LabelEncoder()
data['mainroad']=lb.fit_transform(data['mainroad'])
```

```python
data
```

```python
data.head()
```

```python
data['guestroom']=lb.fit_transform(data['guestroom'])
data['basement']=lb.fit_transform(data['basement'])
data['hotwaterheating']=lb.fit_transform(data['hotwaterheating'])
data['airconditioning']=lb.fit_transform(data['airconditioning'])
data['prefarea']=lb.fit_transform(data['prefarea'])
data['furnishingstatus']=lb.fit_transform(data['furnishingstatus'])
```

```python
data
```

```python
x=data.drop(columns=['price'])
y=data['price']
```

```python
y
```

```python
y=y.values.reshape(-1,1)
s=MinMaxScaler()
x=s.fit_transform(x)
x=s.fit_transform(y)
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model_predict=model.predict(x_train)
```

```python
score=model.score(x_train,y_train)
```

```python
score
```

```python
plt.scatter(y_train,model_predict)
```

```python
score=metrics.mean_absolute_error(y_train, model_predict)
score
```

```python
model_predict=model.predict(x_test)
```

```python
score=model.score(x_test,y_test)
score
```

```python
score=metrics.mean_absolute_error(y_test, model_predict)
score
```
